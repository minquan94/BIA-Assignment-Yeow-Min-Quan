```python
import numpy as np
import pandas as pd
import seaborn as sns
import cufflinks as cf
```

```python
rating_final_df = pd.read_csv('rating_final.csv')
print(rating_final_df.dtypes)
```

```python
userprofile_df = pd.read_csv('userprofile.csv')
print(userprofile_df.dtypes)
```

```python
df = rating_final_df.merge(
    userprofile_df, 
    on=('userID'), 
    how='inner')
df.head()
```

```python
df['rating'] = df['rating'] * (5/2)
df['food_rating'] = df['food_rating'] * (5/2)
df['service_rating'] = df['service_rating'] * (5/2)
df.head()
```

```python
df.weight.unique()
```

```python
columns = list(df)
columns = [i for i in columns if i not in (
    'placeID',
    'userID',
    'food_rating',
    'service_rating',
    'rating')]
df
```

```python
agg_groupby = df.groupby(columns, as_index=False).agg({
    'rating':'mean',
    'food_rating':'mean',
    'service_rating': 'mean',
}).rename(index=str, columns={
    'rating': 'rating_mean',
    'food_rating': 'food_rating_mean',
    'service_rating': 'service_rating_mean'})
agg_groupby
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.lmplot( x="weight", y="rating_mean",data=agg_groupby)


sns.plt.show()
```
